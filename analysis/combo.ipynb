```python
librispeech['dataset'] = 'LS'
ljspeech['dataset'] = 'LJS'
cv['dataset'] = 'CV'
l2['dataset'] = 'L2'
timit['dataset'] = 'TIMIT'

librispeech['phonemes'] = librispeech['gruut_phonemes']
ljspeech['phonemes'] = ljspeech['gruut_phonemes']
cv['phonemes'] = cv['word_phonemes']
l2['phonemes'] = l2['gruut_phonemes']
timit['phonemes'] = timit['words_ipa_mapped_joined']

for dataset in [librispeech, ljspeech, cv, l2, timit]:
    dataset['duration'] = dataset['duration'] / 60

all_data = pd.concat([librispeech, ljspeech, cv, l2, timit], ignore_index=True)
all_data = all_data[['audio_path', 'sentence', 'phonemes', 'duration', 'dataset']]

all_data
```

```python
def select_subset(dataset, total_duration):
    subset = dataset.sample(frac=1, random_state=42)
    subset = subset[subset['duration'].cumsum() <= total_duration]
    return subset

timit_subset = timit  # Весь TIMIT
l2_subset = select_subset(l2, 6)  # 6 часа
ljs_subset = select_subset(ljspeech, 1.35)  # 1.35 часа
ls_subset = select_subset(librispeech, 1.35)  # 1.35 часа
cv_subset = select_subset(cv, 2.7)  # 2.7 часа

subset_data = pd.concat([timit_subset, l2_subset, ljs_subset, ls_subset, cv_subset], ignore_index=True)
subset_data = subset_data[['audio_path', 'sentence', 'phonemes', 'duration', 'dataset']]

subset_data
```

```python
from tqdm import tqdm

def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

def convert_to_wav(input_path, output_path, sample_rate=16000):
    """Конвертирует аудиофайл в формат WAV с частотой дискретизации 16 кГц."""
    sound = AudioSegment.from_file(input_path)
    sound = sound.set_frame_rate(sample_rate)
    sound.export(output_path, format="wav")

def process_audio_files(subset_data, base_dir):
    create_directory(base_dir)

    updated_paths = []
    num_files = len(subset_data)
    
    for i, (index, row) in tqdm(enumerate(subset_data.iterrows()), total=num_files):
        original_path = row['audio_path']
        new_filename = f"{i+1}.wav"
        new_path = os.path.join(base_dir, new_filename)

        convert_to_wav(original_path, new_path)
        updated_paths.append(new_path)

    subset_data['audio_path'] = updated_paths

    return subset_data

base_dir = '/home/morph/Desktop/FINAL/processed_dataset'

subset_data = process_audio_files(subset_data, base_dir)

subset_data.to_csv('/home/morph/Desktop/FINAL/updated_FINAL_DATASET.csv', index=False)
```
