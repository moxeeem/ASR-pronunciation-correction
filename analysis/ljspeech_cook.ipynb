```python
import os
import IPython.display as ipd
from pydub.utils import mediainfo
import pandas as pd
import re
from gruut import sentences
```

# Make LJSpeech 1.1 Great Again

```python
df = pd.read_csv('/home/morph/Desktop/FINAL/LJS/LJSpeech-1.1/metadata.csv', sep='|', header=None)
df.head()
```

```python
audio_dir = '/home/morph/Desktop/FINAL/LJS/LJSpeech-1.1/wavs/'
df['audio_path'] = df[0].apply(lambda x: os.path.join(audio_dir, f"{x}.wav"))

df.head()
```

```python
ipd.Audio(df['audio_path'][122])
```

```python
def get_phonemes(text):
    text = re.sub(r'[^\w\s]', '', text)
    phonemes_list = []
    
    for sent in sentences(text, 
                          lang="en-US", 
                          espeak=False, 
                          punctuations=False, 
                          major_breaks=False, 
                          minor_breaks=False):
        for word in sent:
            if word.phonemes:
                phonemes_list.append(''.join(word.phonemes)) 
    
    return ' '.join(phonemes_list)

df['gruut_phonemes'] = df[1].apply(get_phonemes)

df.head()
```

```python
print(df[1][0])
print(df['gruut_phonemes'][0])

ipd.Audio(df['audio_path'][0])
```

```python
df['gruut_phonemes'] = df['gruut_phonemes'].apply(lambda x: re.sub(r'[]', '', x))
```

```python
print(df[1][0])
print(df['gruut_phonemes'][0])

ipd.Audio(df['audio_path'][0])
```

```python
def get_duration_in_minutes(audio_path):
    audio_info = mediainfo(audio_path)
    duration_seconds = float(audio_info['duration'])
    return duration_seconds / 60  #   

df['duration'] = df['audio_path'].apply(get_duration_in_minutes)
```

```python
df['gruut_phonemes'] = df['gruut_phonemes'].apply(lambda x: x.split())
```

```python
final_df = df[['audio_path', 1, 'gruut_phonemes', 'duration']].copy()
final_df.columns = ['audio_path', 'sentence', 'gruut_phonemes', 'duration']

final_df.head()
```

```python
total_duration = df['duration'].sum()

print(f'   : {total_duration:.2f} ')
```
