```python
from gruut import sentences

# 3. Функция для преобразования текста в фонемы с использованием Gruut
def get_phonemes(text):
    # Убираем знаки препинания
    text = re.sub(r'[^\w\s]', '', text)
    
    phonemes_list = []
    
    # Разбиваем текст на предложения и слова, затем получаем фонемы
    for sent in sentences(text, lang="en-us"):
        for word in sent:
            if word.phonemes:  # Если есть фонемы
                phonemes_list.append(''.join(word.phonemes))  # Объединяем фонемы слова без пробелов
    
    # Возвращаем строку: каждое слово разделено пробелом
    return ' '.join(phonemes_list)

# Применяем преобразование для получения фонем
df['gruut_phonemes'] = df[1].apply(get_phonemes)

df.head()
```

```python
# 4. Функция для получения длительности аудиофайла в минутах
def get_duration_in_minutes(audio_path):
    audio_info = mediainfo(audio_path)
    duration_seconds = float(audio_info['duration'])
    return duration_seconds / 60  # Переводим в минуты

# Применяем функцию для получения длительности
df['duration'] = df['audio_path'].apply(get_duration_in_minutes)
```

```python
df['gruut_phonemes'] = df['gruut_phonemes'].apply(lambda x: x.split())
```

```python
final_df = df[['audio_path', 1, 'gruut_phonemes', 'duration']].copy()
final_df.columns = ['audio_path', 'sentence', 'gruut_phonemes', 'duration']

final_df.head()
```

```python
total_duration = df['duration'].sum()

print(f'Общая длительность всего датасета: {total_duration:.2f} минут')
```
